```python
fig, ax = plt.subplots(figsize=(14,8), dpi=200)

ax.plot(df_tesla["MONTH"], df_tesla["TSLA_USD_CLOSE"], color="red", linewidth=2)

ax.set(title = "Tesla stock price against the Tesla search volume")

#edit on y axis
ax.set_ylabel("Tesla stock price", c="red", fontsize=14)
ax.set_ylim([0, 600])

#edit on x axis
ax.set_xlabel("year", fontsize=14)
ax.set_xlim([df_tesla.MONTH.min(), df_tesla.MONTH.max()])
ax.xaxis.set_minor_locator(MonthLocator())

ax.tick_params("both", labelsize=14)

ax1 = ax.twinx()

ax1.plot(df_tesla["MONTH"], df_tesla["TSLA_WEB_SEARCH"], color="blue", linewidth=2)

ax1.set_ylabel("Tesla search volume", c="blue", fontsize=14)
ax1.tick_params("y", labelsize=14)

plt.show()
```

- spike of interest in the middle of the chart - March 2016. This was when the Tesla Model 3 was unveiled. Also, we can clearly see that the most recent spikes in search coincide, not with the release of a new car, but the roaring stock price for the company!

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate. 
1. Change the title to: Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate <br>
2. Change the y-axis label to: FRED U/E Rate <br>
3. Change the axis limits <br>
4. Add a grey [grid](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.grid.html) to the chart to better see the years and the U/E rate values. Use dashes for the line style<br> 
5. Can you discern any seasonality in the searches? Is there a pattern? 

```python
df_unemployment.head()
```

```python
fig, ax = plt.subplots(figsize=(14,8), dpi=200)

ax.plot(df_unemployment['MONTH'], df_unemployment['UE_BENEFITS_WEB_SEARCH'], c='orange')
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment Search Score", color='orange')

ax.xaxis.grid(True, color='#ddddee', linestyle='--')
ax.xaxis.set_minor_locator(MonthLocator())

ax1 = ax.twinx()


ax1.plot(df_unemployment['MONTH'], df_unemployment['UNRATE'], c='green', linestyle='--')
ax1.set_ylabel("FRED U/E Rate", color='green')

ax1.yaxis.grid(True, color='#ddddee', linestyle='--')

plt.show()
```

- the vertical dashed lines line up with spikes in searches for "Unemployment benefits". Many of the spikes are at year-end - in December. This clearly shows that there is seasonality in the job market.
  
- We see that the financial crisis in 2007/2008 caused a massive spike in unemployment. It took around 10 years (2007-2017) for the unemployment to reach the same level it had before the crisis.

-  the big spike in searches for Unemployment benefits at the end of 2013 was not accompanied by a big increase in the unemployment rate. Something else must have been going on around that time.

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
rolling_UE_data = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()
```

```python
fig, ax = plt.subplots(figsize=(14,8), dpi=200)

ax.plot(df_unemployment['MONTH'], rolling_UE_data['UE_BENEFITS_WEB_SEARCH'], c='skyblue')
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment Search Score", color='skyblue')

ax.xaxis.grid(True, color='#ddddee', linestyle='--')
ax.xaxis.set_minor_locator(MonthLocator())

ax1 = ax.twinx()


ax1.plot(df_unemployment['MONTH'], rolling_UE_data['UNRATE'], c='purple', linestyle='--')
ax1.set_ylabel("FRED U/E Rate", color='purple')

ax1.yaxis.grid(True, color='#ddddee', linestyle='--')

plt.show()
```
